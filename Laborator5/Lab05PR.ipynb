### **1. Logistic Regression**

#### **TEORIE**

* Regresia logistică devine o tehnică de clasificare numai atunci când este adus în imagine un prag de decizie. Stabilirea valorii prag este un aspect foarte important al regresiei logistice și depinde de problema de clasificare în sine.

* Decizia privind valoarea pragului este afectată în mare măsură de valorile preciziei și recall-ului. În mod ideal, dorim că atât precizia, cât și recall-ul să fie 1, dar acest lucru este rareori cazul.

* Pe baza numărului de categorii, regresia logistică poate fi clasificată astfel:
 
  * **binom**: variabilă țintă poate avea doar 2 tipuri posibile: „0” sau „1” care poate reprezenta „câștig” vs „pierdere”, „trece” vs „eșec”, „mort” vs „viu”, etc.
  * **multinomială**: variabilă țintă poate avea 3 sau mai multe tipuri posibile care nu sunt ordonate (adică tipurile nu au semnificație cantitativă), cum ar fi „boală A” vs „boală B” vs „boală C”.
  * **ordinal**: se ocupă de variabile țintă cu categorii ordonate. De exemplu, un scor de test poate fi clasificat că: „foarte slab”, „slab”, „bun”, „foarte bine”. Aici, fiecărei categorii i se poate acordă un scor de genul 0, 1, 2, 3.

* **Linear Regression**: y = $w_{0}$ + $w_{1}$ * X

* **Sigmoid function**: p = $\frac{1}{1 + \frac{1}{(e^y)}}$

* **Logistic Regression**: ln($\frac{p}{1-p}$) =  $w_{0}$ + $w_{1}$ * X

 - p = probability

* **EXEMPLU**: Poate fi folosită pentru a prezice dacă un candidat politic va câștiga sau va pierde alegerile sau dacă un elev de liceu va fi admis sau nu la o anumită facultate. Aceste rezultate binare permit decizii simple între două alternative.

```python
import pandas as pd

dataset = pd.read_csv('../Laborator4/Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)

logisticRegressionScore =  accuracy_score(y_test, y_pred)
print("logisticRegressionScore = " + str(logisticRegressionScore))
```

### **2. Decision Tree Classification**

#### **TEORIE**


* Un arbore de decizie este un tip de învățare automată supravegheată folosit pentru a clasifică sau a face predicții pe baza modului în care s-a răspuns la un set anterior de întrebări. Modelul este o formă de învățare supravegheată, ceea ce înseamnă că modelul este antrenat și testat pe un set de date care conține categorizarea dorită.

* Arborele de decizie poate să nu ofere întotdeauna un răspuns sau o decizie clară. În schimb, poate prezenta opțiuni pentru ca cercetătorul de date să poată lua singur o decizie informată. Arborele de decizie imită gândirea umană, așa că este, în general, ușor pentru oamenii de știință să înțeleagă și să interpreteze rezultatele.

* Termeni cheie ai unui arbore de decizie.

  * **Root node**: baza arborelui de decizie.
  * **Splitting**: Procesul de împărțire a unui nod în mai multe sub-noduri.
  * **Decision node**: Când un sub-nod este împărțit în continuare în sub-noduri suplimentare.
  * **Leaf node**: Când un sub-nod nu se împarte în continuare în sub-noduri suplimentare; reprezintă rezultate posibile.
  * **Pruning**: Procesul de eliminare a sub-nodurilor unui arbore de decizie.
  * **Branch**: o subsecțiune a arborelui de decizie constând din mai multe noduri.

* Arborele de decizie folosește reprezentarea arborelui pentru a rezolva problema în care fiecare nod frunză corespunde unei etichete de clasă și atributele sunt reprezentate pe nodul intern al arborelui.

* În Decision Tree, principala provocare este identificarea atributului pentru nodul rădăcină în fiecare nivel. Acest proces este cunoscut sub numele de selecție de atribute. Avem două măsuri populare de selecție a atributelor:

  * Information Gain
  * Gini Index

### **3. Support Vector Machine (SVM)**

#### **TEORIE**


* Support Vector Machine (SVM) este un algoritm de învățare automată supravegheat folosit atât pentru clasificare, cât și pentru regresie. Deși rezolva și probleme de regresie, este cel mai potrivit pentru clasificare. Obiectivul algoritmului SVM este de a găși un hiperplan într-un spațiu N-dimensional care clasifică distinct punctele de date. Dimensiunea hiperplanului depinde de numărul de caracteristici. Dacă numărul de caracteristici de intrare este două, atunci hiperplanul este doar o linie. Dacă numărul de caracteristici de intrare este trei, atunci hiperplanul devine un plan 2-D. Devine dificil de imaginat când numărul de caracteristici depășește trei.

* O alegere rezonabilă că cel mai bun hiperplan este cea care reprezintă cea mai mare separare sau marjă între cele două clase.

* Algoritmul SVM are caracteristicile de a ignoră valorile aberante și de a găsi cel mai bun hiperplan care maximizează marja. SVM este robust la valori aberante.

* Acesta este unul dintre motivele pentru care folosim SVM-urile în învățarea automată. Poate gestiona atât clasificarea, cât și regresia pe date liniare și neliniare.

* Avantajele SVM:

  * Eficient în cazuri de dimensiuni mari.
  * Memoria sa este eficientă, deoarece utilizează un subset de puncte de antrenament în funcția de decizie numită vectori suport.
  * Pot fi specificate diferite funcții ale nucleului pentru funcțiile de decizie și este posibil să se specifice nuclee personalizate.

### **4. Ensemble Models: Random Forest Classification**

#### **TEORIE**

* Random Forest este un algoritm de învățare automată supravegheat care este utilizat pe scară largă în problemele de clasificare și regresie. Acesta construiește arbori de decizie pe diferite eșantioane și le ia votul majoritar pentru clasificare și medie în cazul regresiei.

* Una dintre cele mai importante caracteristici ale algoritmului Random Forest  este că poate gestiona setul de date care conține variabile continue că în cazul regresiei și variabilele categoriale că în cazul clasificării. Are rezultate mai bune pentru probleme de clasificare.

* **Ensemble** înseamnă pur și simplu combinarea mai multor modele. Astfel, o colecție de modele este folosită pentru a face predicții, mai degrabă decât un model individual.

* **Ensemble** folosește două tipuri de metode:

  * **Bagging** – Creează un subset de antrenament diferit din datele eșantionului de antrenament cu înlocuire, iar rezultatul final se bazează pe votul majoritar. De exemplu, **Random Forest**.

  * **Boosting** – Combină weak learners în strong learners prin crearea unor modele secvențiale astfel încât modelul final să aibă cea mai mare acuratețe. De exemplu, **ADA BOOST, XG BOOST**.
  


4661536426211ba43ea612c8e1a6a1ed45507.png

* **Bagging**
  * **Bagging**, cunoscut și sub denumirea de **Bootstrap Aggregation**, este tehnică de ensemble folosită de Random Forest. Bagging alege un eșantion aleatoriu din setul de date. Prin urmare, fiecare model este generat din eșantioanele (Bootstrap Samples) furnizate de datele originale cu înlocuire cunoscută sub numele de eșantionare pe rând. Acest pas de eșantionare de rând cu înlocuire se numește **bootstrap**. Acum fiecare model este antrenat independent, ceea ce generează rezultate. Rezultatul final se bazează pe votul majoritar după combinarea rezultatelor tuturor modelelor. Acest pas care implică combinarea tuturor rezultatelor și generarea de rezultate bazate pe votul majoritar este cunoscut sub numele de **agregare**.



77772440px-Ensemble_Bagging.svg.png

* Pași implicați în algoritmul Random Forest:

  * **Pasul 1**: Sunt luate n un număr de înregistrări aleatoare din setul de date având k număr de înregistrări.

  * **Pasul 2**: Arbori de decizie individuali sunt construiți pentru fiecare probă.

  * **Pasul 3**: Fiecare arbore de decizie va genera o ieșire.

  * **Pasul 4**: Rezultatul final este considerat pe baza votului majoritar sau a mediei pentru clasificare și, respectiv, regresie.

### **5. Ensemble Models: XGBoost**

#### **TEORIE**

* **XGBoost** este un algoritm de învățare automată, bazat pe arborele de decizie, care utilizează gradient boosting framework. În problemele de predicție care implică date nestructurate (imagini, text etc.), rețelele neuronale artificiale tind să depășească toți ceilalți algoritmi sau framework-uri. Cu toate acestea, atunci când vine vorba de date structurate/tabulare de dimensiuni mici și medii, algoritmii bazați pe arborele de decizie sunt considerați cei mai buni din clasă în acest moment.

* **Evoluția algoritmilor bazați pe arbori de decizie**:

  1. **Arborele de decizie**
  2. **Bagging**
  3. **Random Forest**
  4. **Boosting**
  5. **Gradient Boosting**
  6. **XGBoost**: Gândiți-vă la XGBoost ca la o Gradient Boosting pe „steroizi”. Este o combinație perfectă de tehnici de optimizare software și hardware pentru a obține rezultate superioare folosind mai puține resurse de calcul în cel mai scurt timp.



1_LDfs0CPpxBrJTK8b-f1DWg.png

* **Gradient Boosting**
  * Este un algoritm popular de amplificare. În acest algoritm, fiecare predictor corectează eroarea predecesorului său. Spre deosebire de **Adaboost**, ponderile instanțelor de antrenament nu sunt modificate, în schimb, fiecare predictor este antrenat folosind erorile reziduale ale predecesorului că etichete.

* **XGBoost**
  * XGBoost este o implementare a arborilor de decizie **Gradient Boost**.
  * În acest algoritm, arborii de decizie sunt creați în formă secvențială. Greutățile joacă un rol important în XGBoost. Ponderile sunt atribuite tuturor variabilelor independente care sunt apoi introduse în arborele de decizie care prezice rezultatele. Ponderea variabilelor prezise greșit de arbore este crescută și aceste variabile sunt apoi introduse în al doilea arbore de decizie. Acești clasificatori/predictori individuali apoi se adună pentru a oferi un model puternic și mai precis. Poate funcționa pe probleme de regresie, clasificare, clasare și predicții definite de utilizator.

1_FLshv-wVDfu-i54OqvZdHg.png

* **Optimizarea sistemului**:

  * **Paralelizare**: XGBoost abordează procesul de construire secvențială a arborilor folosind implementarea paralelizată. Acest lucru este posibil datorită naturii interschimbabile a buclelor utilizate pentru construirea learnerilor de bază; buclă exterioară care enumeră nodurile frunzelor unui arbore și a două buclă interioară care calculează caracteristicile. Această imbricare a buclelor limitează paralelizarea deoarece fără a finaliza buclă interioară (mai solicitantă din punct de vedere computațional pentru cele două), buclă exterioară nu poate fi pornită. Prin urmare, pentru a îmbunătăți timpul de rulare, ordinea buclelor este schimbată folosind inițializarea printr-o scanare globală a tuturor instanțelor și sortarea folosind fire paralele. Acest comutator îmbunătățește performanța algoritmică prin compensarea oricăror costuri de paralelizare în calcul.

  * **Tree Pruning**: Criteriul de oprire pentru despicarea copacilor este lacom în natură și depinde de criteriul pierderii negative în punctul de despicare. XGBoost folosește parametrul „max_depth” așa cum este specificat în loc de criteriu mai întâi și începe tăierea copacilor înapoi. Această abordare „în profunzime” îmbunătățește semnificativ performanța de calcul.

  * **Optimizare hardware**: Acest algoritm a fost conceput pentru a utiliza eficient resursele hardware. Acest lucru se realizează prin conștientizarea memoriei cache prin alocarea de buffer-uri interne în fiecare fir pentru a stoca statistici de gradient. Îmbunătățiri suplimentare, cum ar fi calculul „out-of-core” optimizează spațiul disponibil pe disc, în timp ce gestionează cadre mari de date care nu se potrivesc în memorie.

* **Îmbunătățiri algoritmice**:

  * **Regularizare**: penalizează modelele mai complexe atât prin regularizarea LASSO (L1), cât și prin regularizarea Ridge (L2), pentru a preveni overfitting-ul.

  * **Conștientizarea dispersității**: XGBoost admite în mod natural funcții rare pentru întrări, „învățând” automat cea mai bună valoare lipsă în funcție de pierderea antrenamentului și gestionează mai eficient diferitele tipuri de modele de dispersie în date.

  * **Weighted Quantile Sketch**: XGBoost folosește algoritmul ponderat Quantile Sketch distribuit pentru a găși în mod eficient punctele de împărțire optime între seturile de date ponderate.

  * **Cross-validation**: algoritmul vine cu o metodă de validare încrucișată încorporată la fiecare iterație, eliminând nevoia de a programa în mod explicit această căutare și de a specifică numărul exact de iterații de stimulare necesare într-o singură rulare.

### **7.** *(OPTIONAL)* **Kernel SVM**

#### **TEORIE**

* Un nucleu este o funcție folosită în SVM pentru a ajută la rezolvarea problemelor. Ele oferă comenzi rapide pentru a evita calculele complexe. Lucrul uimitor despre nucleu este că putem merge la dimensiuni mai mari și putem efectua calcule netede cu ajutorul acestuia.

* Putem merge până la un număr infinit de dimensiuni folosind nuclee.

* Uneori, nu putem avea un hiperplan pentru anumite probleme. Această problemă apare atunci când urcăm la dimensiuni mai mari și încercăm să formăm un hiperplan.

* Un nucleu ajută la formarea hiperplanului în dimensiunea superioară fără a crește complexitatea.


* Kernel-urile sunt o modalitate de a rezolva probleme neliniare cu ajutorul clasificatorilor liniari. Această este cunoscută că metodă trucului nucleului. Funcțiile nucleului sunt utilizate că parametri în codurile SVM. Ele ajută la determinarea formei hiperplanului și a limitei de decizie.

* Putem seta valoarea parametrului kernelului în codul SVM.

* Valoarea poate fi orice tip de nucleu de la liniar la polinom. Dacă valoarea nucleului este liniară, atunci granița de decizie ar fi liniară și bidimensională. Aceste funcții ale nucleului ajută, de asemenea, la stabilirea limitelor de decizie pentru dimensiuni mai mari.

* Nu trebuie să facem calcule complexe. Funcțiile nucleului fac toată muncă grea. Trebuie doar să dăm intrarea și să folosim nucleul corespunzător. De asemenea, putem rezolva problema supraadaptarii în SVM folosind funcțiile kernelului.

* Tipuri de funcții ale nucleului:

  1. **Linear kernel**
  * $K(x_{i}, x_{j})$ = $x_{i}x_{j} + c$

  2. **RBF kernel**
  *  $K(x_{i}, x_{j})$ = ${e}^{(-Y||x_{i} - x_{j}||)^2}$

  3. **Sigmoid kernel**
  *  $K(x_{i}, x_{j})$ = $tanh(Y(x_{i}x_{j})+ c)$

  4. **Polynomial kernel**
  * $K(x_{i}, x_{j})$ = $(x_{i}x_{j} + c)^{d}$
  * d = gradul polinomului

* **EXEMPLU**:
  * Beneficiul final este că funcția obiectiv pe care o optimizăm pentru a se potrivi cu granița de decizie dimensională superioară include doar produsul punctual al vectorilor caracteristici transformați.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC

classifier_linear = SVC(kernel='linear', random_state=0)
classifier_linear.fit(X_train, y_train)

classifier_poly = SVC(kernel='poly', random_state=0)
classifier_poly.fit(X_train, y_train)

classifier_rbf = SVC(kernel='rbf', random_state=0)
classifier_rbf.fit(X_train, y_train)

classifier_sigmoid = SVC(kernel='sigmoid', random_state=0)
classifier_sigmoid.fit(X_train, y_train)
```

```python
y_pred_linear = classifier_linear.predict(X_test)
y_pred_poly = classifier_poly.predict(X_test)
y_pred_rbf = classifier_rbf.predict(X_test)
y_pred_sigmoid = classifier_sigmoid.predict(X_test)
```

## **1. Clusterizare**

### **K-Means**

#### **TEORIE**

* K-means este un algoritm de grupare bazat pe centroid, în care calculăm distanța dintre fiecare punct de date și un centroid pentru a-l atribui unui cluster. Scopul este de a identifica numărul K de grupuri din setul de date.

* Este un proces iterativ de atribuire a fiecărui punct de date către grupuri și, încet, punctele de date sunt grupate pe baza unor caracteristici similare. Obiectivul este de a minimiza suma distanțelor dintre punctele de date și centroidul clusterului, pentru a identifica grupul corect căruia ar trebui să aparțînă fiecare punct de date.

* Aici, împărțim un spațiu de date în K clustere și atribuim o valoare medie fiecăruia. Punctele de date sunt plasate în clusterele cele mai apropiate de valoarea medie a acelui cluster. Există mai multe valori de distanță disponibile care pot fi utilizate pentru a calcula distanța.





#### **Cum funcționează K-means?** 
* Să luăm un exemplu pentru a înțelege cum funcționează K-means pas cu pas. 
* Algoritmul poate fi împărțit în 4-5 pași:

  1. **Alegerea numărului de clustere**
  * Primul pas este definirea numărului ${K}$ de clustere în care vom grupa datele. Să alegem ${K=3}$.

  2. **Inițializarea centroizilor**
    * Centroidul este centrul unui cluster, dar inițial, centrul exact al punctelor de date va fi necunoscut, așa că selectăm puncte de date aleatorii și le definim ca centroizi pentru fiecare cluster. Vom inițializa 3 centroizi în setul de date.

K-Means-Clustering_7.png

  3. **Atribuiți puncte de date celui mai apropiat cluster**
    * Acum că centroizii sunt inițializați, următorul pas este să atribuiți punctele de date ${X_n}$ celui mai apropiat centroid ${C_k}$ al lor.

    * În acest pas, vom calcula mai întâi distanța dintre punctul de date ${X}$ și centroidul ${C}$ folosind metrica Distanța Euclidiană.

    * Și apoi alegeți clusterul pentru punctele de date unde distanța dintre punctul de date și centroid este minimă.

1.png2.png3.png

  4. **Reinițializați centroizii**
    * În continuare, vom reinițializa centroizii calculând media tuturor punctelor de date ale acelui cluster.

4.png

  5. **Repetați pașii 3 și 4**
    * Vom continua să repetăm pașii 3 și 4 până când vom avea centroizi optimi și alocarea punctelor de date pentru clusterele corecte nu se mai schimbă.

6.png

* **OBS**: Pentru a alege numarul potrivit de clustere la inceput folosim **WCSS (Within Cluster Sum of Squares)** =>

 - **Within Cluster Sum of Squares ($WCSS$)** = ${{\sum_{j=1}^{number Of Clusters} \sum_{i=1}^{P_{i} ⊂ Cluster_{j}} (distance(P_{i}, C_{j}))^2 }}$
   
* **EXEMPLU**:
  * detectarea spam-ului
  * analiza sentimentelor
  * previziunea examenelor pe baza de scor

```python
import pandas as pd

dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values

print(dataset.head())
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

### **Task 2** - Importul setului de date

```python
dataset = pd.read_csv('Iris.csv')
```

### **Task 3** - Preprocesarea datelor

```python
X = dataset.iloc[:, [1, 2, 3, 4]].values

dataset
```

### **Task 4** - Aflarea numarului potrivit de centroizi

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

### **Task 5** - Folosirea K-Means pe X

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

print(y_kmeans)
print(dataset['Species'])
```

### **Task 6** - Plot K-Means

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', label = 'Centroids')
plt.title('Clusters of leaves')
plt.xlabel('Leaf Dimensions')
plt.ylabel('Type of leaf')
plt.legend()
plt.show()
```
