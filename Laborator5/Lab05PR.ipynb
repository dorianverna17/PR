### **1. Logistic Regression**

#### **TEORIE**

* Regresia logistic devine o tehnic de clasificare numai atunci cnd este adus n imagine un prag de decizie. Stabilirea valorii prag este un aspect foarte important al regresiei logistice i depinde de problema de clasificare n sine.

* Decizia privind valoarea pragului este afectat n mare msur de valorile preciziei i recall-ului. n mod ideal, dorim c att precizia, ct i recall-ul s fie 1, dar acest lucru este rareori cazul.

* Pe baza numrului de categorii, regresia logistic poate fi clasificat astfel:
 
  * **binom**: variabil int poate avea doar 2 tipuri posibile: 0 sau 1 care poate reprezenta ctig vs pierdere, trece vs eec, mort vs viu, etc.
  * **multinomial**: variabil int poate avea 3 sau mai multe tipuri posibile care nu sunt ordonate (adic tipurile nu au semnificaie cantitativ), cum ar fi boal A vs boal B vs boal C.
  * **ordinal**: se ocup de variabile int cu categorii ordonate. De exemplu, un scor de test poate fi clasificat c: foarte slab, slab, bun, foarte bine. Aici, fiecrei categorii i se poate acord un scor de genul 0, 1, 2, 3.

* **Linear Regression**: y = $w_{0}$ + $w_{1}$ * X

* **Sigmoid function**: p = $\frac{1}{1 + \frac{1}{(e^y)}}$

* **Logistic Regression**: ln($\frac{p}{1-p}$) =  $w_{0}$ + $w_{1}$ * X

 - p = probability

* **EXEMPLU**: Poate fi folosit pentru a prezice dac un candidat politic va ctiga sau va pierde alegerile sau dac un elev de liceu va fi admis sau nu la o anumit facultate. Aceste rezultate binare permit decizii simple ntre dou alternative.

```python
import pandas as pd

dataset = pd.read_csv('../Laborator4/Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
y_pred = logisticRegressionClassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)

logisticRegressionScore =  accuracy_score(y_test, y_pred)
print("logisticRegressionScore = " + str(logisticRegressionScore))
```

### **2. Decision Tree Classification**

#### **TEORIE**


* Un arbore de decizie este un tip de nvare automat supravegheat folosit pentru a clasific sau a face predicii pe baza modului n care s-a rspuns la un set anterior de ntrebri. Modelul este o form de nvare supravegheat, ceea ce nseamn c modelul este antrenat i testat pe un set de date care conine categorizarea dorit.

* Arborele de decizie poate s nu ofere ntotdeauna un rspuns sau o decizie clar. n schimb, poate prezenta opiuni pentru ca cercettorul de date s poat lua singur o decizie informat. Arborele de decizie imit gndirea uman, aa c este, n general, uor pentru oamenii de tiin s neleag i s interpreteze rezultatele.

* Termeni cheie ai unui arbore de decizie.

  * **Root node**: baza arborelui de decizie.
  * **Splitting**: Procesul de mprire a unui nod n mai multe sub-noduri.
  * **Decision node**: Cnd un sub-nod este mprit n continuare n sub-noduri suplimentare.
  * **Leaf node**: Cnd un sub-nod nu se mparte n continuare n sub-noduri suplimentare; reprezint rezultate posibile.
  * **Pruning**: Procesul de eliminare a sub-nodurilor unui arbore de decizie.
  * **Branch**: o subseciune a arborelui de decizie constnd din mai multe noduri.

* Arborele de decizie folosete reprezentarea arborelui pentru a rezolva problema n care fiecare nod frunz corespunde unei etichete de clas i atributele sunt reprezentate pe nodul intern al arborelui.

* n Decision Tree, principala provocare este identificarea atributului pentru nodul rdcin n fiecare nivel. Acest proces este cunoscut sub numele de selecie de atribute. Avem dou msuri populare de selecie a atributelor:

  * Information Gain
  * Gini Index

### **3. Support Vector Machine (SVM)**

#### **TEORIE**


* Support Vector Machine (SVM) este un algoritm de nvare automat supravegheat folosit att pentru clasificare, ct i pentru regresie. Dei rezolva i probleme de regresie, este cel mai potrivit pentru clasificare. Obiectivul algoritmului SVM este de a gi un hiperplan ntr-un spaiu N-dimensional care clasific distinct punctele de date. Dimensiunea hiperplanului depinde de numrul de caracteristici. Dac numrul de caracteristici de intrare este dou, atunci hiperplanul este doar o linie. Dac numrul de caracteristici de intrare este trei, atunci hiperplanul devine un plan 2-D. Devine dificil de imaginat cnd numrul de caracteristici depete trei.

* O alegere rezonabil c cel mai bun hiperplan este cea care reprezint cea mai mare separare sau marj ntre cele dou clase.

* Algoritmul SVM are caracteristicile de a ignor valorile aberante i de a gsi cel mai bun hiperplan care maximizeaz marja. SVM este robust la valori aberante.

* Acesta este unul dintre motivele pentru care folosim SVM-urile n nvarea automat. Poate gestiona att clasificarea, ct i regresia pe date liniare i neliniare.

* Avantajele SVM:

  * Eficient n cazuri de dimensiuni mari.
  * Memoria sa este eficient, deoarece utilizeaz un subset de puncte de antrenament n funcia de decizie numit vectori suport.
  * Pot fi specificate diferite funcii ale nucleului pentru funciile de decizie i este posibil s se specifice nuclee personalizate.

### **4. Ensemble Models: Random Forest Classification**

#### **TEORIE**

* Random Forest este un algoritm de nvare automat supravegheat care este utilizat pe scar larg n problemele de clasificare i regresie. Acesta construiete arbori de decizie pe diferite eantioane i le ia votul majoritar pentru clasificare i medie n cazul regresiei.

* Una dintre cele mai importante caracteristici ale algoritmului Random Forest  este c poate gestiona setul de date care conine variabile continue c n cazul regresiei i variabilele categoriale c n cazul clasificrii. Are rezultate mai bune pentru probleme de clasificare.

* **Ensemble** nseamn pur i simplu combinarea mai multor modele. Astfel, o colecie de modele este folosit pentru a face predicii, mai degrab dect un model individual.

* **Ensemble** folosete dou tipuri de metode:

  * **Bagging**  Creeaz un subset de antrenament diferit din datele eantionului de antrenament cu nlocuire, iar rezultatul final se bazeaz pe votul majoritar. De exemplu, **Random Forest**.

  * **Boosting**  Combin weak learners n strong learners prin crearea unor modele secveniale astfel nct modelul final s aib cea mai mare acuratee. De exemplu, **ADA BOOST, XG BOOST**.
  


4661536426211ba43ea612c8e1a6a1ed45507.png

* **Bagging**
  * **Bagging**, cunoscut i sub denumirea de **Bootstrap Aggregation**, este tehnic de ensemble folosit de Random Forest. Bagging alege un eantion aleatoriu din setul de date. Prin urmare, fiecare model este generat din eantioanele (Bootstrap Samples) furnizate de datele originale cu nlocuire cunoscut sub numele de eantionare pe rnd. Acest pas de eantionare de rnd cu nlocuire se numete **bootstrap**. Acum fiecare model este antrenat independent, ceea ce genereaz rezultate. Rezultatul final se bazeaz pe votul majoritar dup combinarea rezultatelor tuturor modelelor. Acest pas care implic combinarea tuturor rezultatelor i generarea de rezultate bazate pe votul majoritar este cunoscut sub numele de **agregare**.

* Pai implicai n algoritmul Random Forest:

  * **Pasul 1**: Sunt luate n un numr de nregistrri aleatoare din setul de date avnd k numr de nregistrri.

  * **Pasul 2**: Arbori de decizie individuali sunt construii pentru fiecare prob.

  * **Pasul 3**: Fiecare arbore de decizie va genera o ieire.

  * **Pasul 4**: Rezultatul final este considerat pe baza votului majoritar sau a mediei pentru clasificare i, respectiv, regresie.

### **7.** *(OPTIONAL)* **Kernel SVM**

#### **TEORIE**

* Un nucleu este o funcie folosit n SVM pentru a ajut la rezolvarea problemelor. Ele ofer comenzi rapide pentru a evita calculele complexe. Lucrul uimitor despre nucleu este c putem merge la dimensiuni mai mari i putem efectua calcule netede cu ajutorul acestuia.

* Putem merge pn la un numr infinit de dimensiuni folosind nuclee.

* Uneori, nu putem avea un hiperplan pentru anumite probleme. Aceast problem apare atunci cnd urcm la dimensiuni mai mari i ncercm s formm un hiperplan.

* Un nucleu ajut la formarea hiperplanului n dimensiunea superioar fr a crete complexitatea.


* Kernel-urile sunt o modalitate de a rezolva probleme neliniare cu ajutorul clasificatorilor liniari. Aceast este cunoscut c metod trucului nucleului. Funciile nucleului sunt utilizate c parametri n codurile SVM. Ele ajut la determinarea formei hiperplanului i a limitei de decizie.

* Putem seta valoarea parametrului kernelului n codul SVM.

* Valoarea poate fi orice tip de nucleu de la liniar la polinom. Dac valoarea nucleului este liniar, atunci grania de decizie ar fi liniar i bidimensional. Aceste funcii ale nucleului ajut, de asemenea, la stabilirea limitelor de decizie pentru dimensiuni mai mari.

* Nu trebuie s facem calcule complexe. Funciile nucleului fac toat munc grea. Trebuie doar s dm intrarea i s folosim nucleul corespunztor. De asemenea, putem rezolva problema supraadaptarii n SVM folosind funciile kernelului.

* Tipuri de funcii ale nucleului:

  1. **Linear kernel**
  * $K(x_{i}, x_{j})$ = $x_{i}x_{j} + c$

  2. **RBF kernel**
  *  $K(x_{i}, x_{j})$ = ${e}^{(-Y||x_{i} - x_{j}||)^2}$

  3. **Sigmoid kernel**
  *  $K(x_{i}, x_{j})$ = $tanh(Y(x_{i}x_{j})+ c)$

  4. **Polynomial kernel**
  * $K(x_{i}, x_{j})$ = $(x_{i}x_{j} + c)^{d}$
  * d = gradul polinomului

* **EXEMPLU**:
  * Beneficiul final este c funcia obiectiv pe care o optimizm pentru a se potrivi cu grania de decizie dimensional superioar include doar produsul punctual al vectorilor caracteristici transformai.

## **1. Clusterizare**

### **K-Means**

#### **TEORIE**

* K-means este un algoritm de grupare bazat pe centroid, n care calculm distana dintre fiecare punct de date i un centroid pentru a-l atribui unui cluster. Scopul este de a identifica numrul K de grupuri din setul de date.

* Este un proces iterativ de atribuire a fiecrui punct de date ctre grupuri i, ncet, punctele de date sunt grupate pe baza unor caracteristici similare. Obiectivul este de a minimiza suma distanelor dintre punctele de date i centroidul clusterului, pentru a identifica grupul corect cruia ar trebui s aparn fiecare punct de date.

* Aici, mprim un spaiu de date n K clustere i atribuim o valoare medie fiecruia. Punctele de date sunt plasate n clusterele cele mai apropiate de valoarea medie a acelui cluster. Exist mai multe valori de distan disponibile care pot fi utilizate pentru a calcula distana.





#### **Cum funcioneaz K-means?** 
* S lum un exemplu pentru a nelege cum funcioneaz K-means pas cu pas. 
* Algoritmul poate fi mprit n 4-5 pai:

  1. **Alegerea numrului de clustere**
  * Primul pas este definirea numrului ${K}$ de clustere n care vom grupa datele. S alegem ${K=3}$.

  2. **Iniializarea centroizilor**
    * Centroidul este centrul unui cluster, dar iniial, centrul exact al punctelor de date va fi necunoscut, aa c selectm puncte de date aleatorii i le definim ca centroizi pentru fiecare cluster. Vom iniializa 3 centroizi n setul de date.

K-Means-Clustering_7.png

  3. **Atribuii puncte de date celui mai apropiat cluster**
    * Acum c centroizii sunt iniializai, urmtorul pas este s atribuii punctele de date ${X_n}$ celui mai apropiat centroid ${C_k}$ al lor.

    * n acest pas, vom calcula mai nti distana dintre punctul de date ${X}$ i centroidul ${C}$ folosind metrica Distana Euclidian.

    * i apoi alegei clusterul pentru punctele de date unde distana dintre punctul de date i centroid este minim.

1.png2.png3.png

  4. **Reiniializai centroizii**
    * n continuare, vom reiniializa centroizii calculnd media tuturor punctelor de date ale acelui cluster.

4.png

  5. **Repetai paii 3 i 4**
    * Vom continua s repetm paii 3 i 4 pn cnd vom avea centroizi optimi i alocarea punctelor de date pentru clusterele corecte nu se mai schimb.

6.png

* **OBS**: Pentru a alege numarul potrivit de clustere la inceput folosim **WCSS (Within Cluster Sum of Squares)** =>

 - **Within Cluster Sum of Squares ($WCSS$)** = ${{\sum_{j=1}^{number Of Clusters} \sum_{i=1}^{P_{i}  Cluster_{j}} (distance(P_{i}, C_{j}))^2 }}$
   
* **EXEMPLU**:
  * detectarea spam-ului
  * analiza sentimentelor
  * previziunea examenelor pe baza de scor

```python
import pandas as pd

dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values

print(dataset.head())
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

### **Task 5** - Folosirea K-Means pe X

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

print(y_kmeans)
print(dataset['Species'])
```

### **Task 6** - Plot K-Means

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', label = 'Centroids')
plt.title('Clusters of leaves')
plt.xlabel('Leaf Dimensions')
plt.ylabel('Type of leaf')
plt.legend()
plt.show()
```
