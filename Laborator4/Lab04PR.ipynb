# **III.** *(OPȚIONAL)* **Preprocesarea datelor**

## **TEORIE**

### **Ce este preprocesarea datelor?**

* Preprocesarea datelor include pașii pe care trebuie să-i urmăm pentru a transformă sau codifică datele, astfel încât să poată fi analizate cu ușurință de către mașînă.

* Ideea principală pentru că un model să fie precis în predicții este că algoritmul să poată interpreta cu ușurință caracteristicile datelor.

* Majoritatea seturilor de date din lumea reală pentru ML sunt susceptibile să aibă date lipsa, inconsecvențe și zgomot din cauza originii lor eterogene.

* Aplicarea algoritmilor de extragere a datelor pe aceste date zgomotoase nu ar da rezultate de calitate, deoarece nu ar reuși să identifice modelele în mod eficient. Preprocesarea datelor este, prin urmare, importantă pentru a îmbunătăți calitatea generală a datelor.

* Valorile duplicate sau lipsă pot oferi o imagine incorectă a statisticilor generale ale datelor.

* Valorile aberante și punctele de date inconsecvențe tind adesea să perturbe învățarea generală a modelului, ceea ce duce la predicții false.


613749410d056eb67ec4b11f_model-building.png

### **Tratarea datelor lipsă**

* Există multe motive pentru care datele ar putea lipsi dintr-un set de date. De exemplu, datele colectate printr-un sondaj pot avea date lipsă din cauza eșecului participanților de a răspunde la unele întrebări, a nu cunoaște răspunsul corect sau că nu doresc să răspundă. Pot lipsi și din cauza erorii făcute în timpul procesului de introducere a datelor.

* Majoritatea modelelor de învățare automată necesită date cu o valoare pentru toate caracteristicile din fiecare observație. În astfel de modele, lipsa datelor poate duce la părtinire în estimarea parametrilor și, de asemenea, poate compromite acuratețea modelelor de ML.

#### **Există mai multe tehnici pe care le folosim pentru a gestiona datele lipsă, printre care menționăm:**

##### 1. **Ștergerea liniei cu valoarea/valorile lipsă** (nerecomandată)



  * Această tehnică funcționează bine pe seturi mari de date cu puține valori lipsă. De exemplu, ștergerea unui rând dintr-un set de date cu sute de observații nu poate afecta calitatea informațiilor setului de date. Cu toate acestea, această tehnică nu este potrivită pentru un set de date care raportează multe valori lipsă. Ștergerea multor rânduri dintr-un set de date duce la pierderea informațiilor.

##### 2. **Media** (nerecomandată)



* În cadrul acestei tehnici, înlocuim valoarea lipsă cu media variabilei în care apare. Avantajul acestei tehnici este că păstrează media și dimensiunea eșantionului. Cu toate acestea, această tehnică are unele dezavantaje serioase.

* Imputarea medie subestimează eroarea standard și nu păstrează corelația dintre variabile. Relația dintre variabile este un aspect esențial al analizei, deoarece obiectivul general al studiului este de a o înțelege mai bine.

* Prin urmare, imputarea mediei nu este o soluție adecvată pentru datele lipsă decât dacă datele lipsesc complet la întâmplare (datele lipsă nu au nicio legătură atât cu datele lipsă, cât și cu valorile observate din setul de date).

##### 3. **Hot Deck** (recomandată)



* În această tehnică, înlocuim valoarea lipsă a observației cu o valoare selectată aleatoriu din toate observațiile din eșantion care are valori similare pentru alte variabile.

* Astfel, această tehnică asigură că valoarea de imputare este selectată numai din intervalul posibil în care valoarea reală ar putea probabil să scadă și este selectată aleatoriu, mai degrabă decât să fie determinată, ceea ce este un aspect esențial pentru o eroare standard corectă.

##### 4. **Cold Deck** (recomandată)



* Înlocuim datele lipsă folosind o valoare aleasă dintre alte variabile cu valori similare de observație în această tehnică. Diferența dintre această tehnică și imputarea Hoț Deck este că procesul de selectare a valorii de imputare nu este randomizat.

##### 5. **Regresie** (recomandată)



* Regresia implică ajustarea unui model de regresie pe o caracteristică cu date lipsă și apoi utilizarea predicțiilor acestui model de regresie pentru a înlocui valorile lipsă din această caracteristică. Această tehnică păstrează relațiile dintre caracteristici și această îi conferă un avantaj semnificativ față de tehnicile simple de inlocuiew, cum ar fi  media și modulul.

### **Tehnici de preprocesare a datelor pentru ML**

#### **1. Redimensionarea datelor**



* Atunci când datele sunt compuse din atribute cu plaje de valori diferite, mulți algoritmi de ML pot beneficia de redimensionarea atributelor pentru că toate să aibă aceeași scară.

* Acest lucru este util pentru algoritmii de optimizare utilizați în nucleul algoritmilor de ML, cum ar fi gradient descent.

* De asemenea, este util pentru algoritmii care ponderează intrările, cum ar fi regresia și rețelele neuronale și algoritmii care folosesc măsuri de distanță, cum ar fi K-Nearest Neighbors.

* Putem redimensiona datele folosind clasa MinMaxScaler.

````python
import pandas
import scipy
import numpy
from sklearn.preprocessing import MinMaxScaler
 
"""
https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database?resource=download

This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. 
The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on 
certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of 
these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.
"""

names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

dataframe = pandas.read_csv("diabetes.csv", names=names)
array = dataframe.values

print("dataframe = ")
print(dataframe.iloc[:10,:])

print("---------------------------------------------------")
print("array = ")
print(array)

# variable independente
X = array[1:,0:8]
# variable dependente
Y = array[1:,8]

print("---------------------------------------------------")
print("X = ")
print(X[:10])

print("---------------------------------------------------")
print("Y = ")
print(Y[:10])
 

scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
 
numpy.set_printoptions(precision=3)

print("---------------------------------------------------")
print("rescaledX = ")
print(rescaledX[0:10,:])
````

* Standardizarea este o tehnică utilă pentru a transformă atribute cu o distribuție gaussiana și diferite medii și abateri standard într-o distribuție gaussiană standard cu o medie de 0 și o abatere standard de 1.

### **Feature Scaling**

**OBS**: Se aplică după împărțirea datelor în antrenare și testare pentru a nu se produce pierdere de informație nouă.

**OBS**: Se folosește pentru că anumite feature-uri să nu fie dominate de altele, le aducem în aceeași plajă de valori. Putem să realizăm acest lurcru prin:

  1. **Standardizare** 

  - **$X_{stand}$** = $\frac{X - mean(X)}{standardDeviation(X)}$ ∈ [-3, +3]
  - **$standardDeviation(X)$** = $\sqrt{\frac{\sum_{i=1}^{n} (x_{i} - x')^2}{n-1}}$
    - x' = media valorilor lui X
    - n  = nr total de valori

  2. **Normalizare** 

  - **$x_{norm}$** = $\frac{x - min(X)}{max{X}-min(X)}$ ∈ [-1, +1]

**OBS** : **Nu** faci feature scaling pe datele de testare.

# **IV. Învățarea Supervizată**

## **1. Regresie**

### **1. Linear Regression**

#### **TEORIE**

* Regresia modelează o valoare de predicție țintă bazată pe variabile independente. Este folosit mai ales pentru a află relația dintre variabile și prognoză.

* Regresia liniară îndeplinește sarcina de a prezice o valoare a variabilei dependențe (y) pe baza unei variabile independente date (x). Deci, această tehnică de regresie descoperă o relație liniară între x (intrare) și y (ieșire).

* **y = $w_{0}$ + $w_{1}$ * $x$**

  - **y** = variabilă dependență
  - **x** = variabilă independența
  - **$w_{1}$** = coeficient
  - **$w_{0}$** = bias

* Când antrenăm modelul – potrivim cea mai bună linie pentru a prezice valoarea lui y pentru o valoare dată a lui x. Modelul obține cea mai bună linie de potrivire de regresie prin găsirea celor mai bune valori **$w_{0}$** și **$w_{1}$**.

* Cum se actualizează valorile **$w_{0}$** și **$w_{1}$** pentru a obține cea mai bună linie de regresie?

* Prin realizarea liniei de regresie cea mai bună, modelul urmărește să prezică valoarea y astfel încât diferența de eroare dintre valoarea prezisă și valoarea adevărată să fie minimă. Deci, este foarte important să actualizăm valorile  **$w_{0}$** și **$w_{1}$**, pentru a ajunge la cea mai bună valoare care să minimizeze eroarea dintre valoarea y prezisă (pred) și valoarea y adevărată (y).

  - **$RMSE$** = $\frac{1}{n}$ * ${\sum_{i=1}^{n} (pred_{i} - y_{i})^2}$

  - **minimize** **$RMSE$**

* Root Mean Squared Error (**$RMSE$**)

### **3. Random Forest Regression**

#### **TEORIE**

* Fiecare arbore de decizie are o varianță mare, dar când le combinăm pe toate împreună în paralel, varianța rezultată este scăzută, deoarece fiecare arbore de decizie este antrenat perfect pe acele date eșantion, și, prin urmare, rezultatul nu depinde de un arbore de decizie, ci de arbori de decizie multipli. În cazul unei probleme de clasificare, rezultatul final este luat prin utilizarea clasificatorului cu vot majoritar. În cazul unei probleme de regresie, rezultatul final este media tuturor rezultatelor. Această parte se numește agregare.

* Random Forest este o ensemble technique capabilă să efectueze atât sarcini de regresie, cât și de clasificare cu ajutorul mai multor arbori de decizie și o tehnică numită **Bootstrap & Aggregation**, cunoscută în mod obișnuit că **bagging**. Ideea de bază din spatele acestui lucru este de a combină mai mulți arbori de decizie în determinarea rezultatului final, mai degrabă decât să se bazeze pe arbori de decizie individuali.
Random Forest are mai mulți arbori de decizie că modele de bază de învățare. Efectuăm aleatoriu eșantionarea rândurilor și eșantionarea caracteristicilor din setul de date, formând seturi de date mostre pentru fiecare model. Această parte se numește **Bootstrap**.

* **EXEMPLU**: Random Forest este folosit atât pentru clasificare, cât și pentru regresie - de exemplu, pentru clasificarea dacă un e-mail este „spam” sau „nu spam”. Random Forest este utilizat în multe industrii diferite, inclusiv bancar, comerț și asistență medicală.

### **4. Multiple Linear Regression**

#### **TEORIE**

* Regresia este o metodă statistică pentru a determina relația dintre caracteristici și o variabilă sau rezultat. Învățarea automată, este utilizată că metodă de modelare predictivă, în care se folosește un algoritm pentru a prognoza rezultate continue. Regresia liniară multiplă, cunoscută adesea sub numele de regresie multiplă, este o metodă statistică care prezice rezultatul unei variabile de răspuns prin combinarea a numeroase variabile explicative. Regresia multiplă este o variantă a regresiei liniare în care este utilizată o singură variabilă explicativă.

*  **y = $w_{0}$ + $w_{1}$ * $x_{1}$ + $w_{2}$ * $x_{2}$ + ... + $w_{n}$ * $x_{n}$**

  - **y** = variabilă dependență
  - **$x_{1}$, $x_{2}$, ... , $x_{n}$** = variabilele independente
  - **$w_{1}$, $w_{2}$, ... , $w_{n}$** = coeficienți
  - **$w_{0}$** = bias

* **EXEMPLU**: Un analist poate dori să știe cum mișcarea pieței afectează prețul ExxonMobil (XOM). În acest caz, ecuația lor liniară va avea valoarea indicelui S&P 500 că variabilă independentă, sau predictor, și prețul XOM că variabilă dependentă.

### **5. Polynomial Regression**

#### **TEORIE**


* Regresia polinomială este o formă de regresie liniară în care relația dintre variabila independentă x și variabila dependentă y este modelată ca un polinom de gradul al n-lea. Regresia polinomială se potrivește unei relații neliniare între valoarea lui x și media condiționată corespunzătoare a lui y, notată $E(y|x)$

* De ce regresia polinomială?
 
  * Există unele relații despre care un cercetător le va ipoteza că sunt curbilinii. În mod clar, astfel de tipuri de cazuri vor include un termen polinom.Verificarea reziduurilor. Dacă încercăm să potrivim un model liniar la datele curbe, un grafic de împrăștiere a reziduurilor (axa Y) pe predictor (axa X) va avea pete cu multe reziduuri pozitive în mijloc. Prin urmare, într-o astfel de situație, nu este potrivit.
  * O ipoteză în analiza de regresie liniară multiplă obișnuită este că toate variabilele independente sunt independente. În modelul de regresie polinomială, această ipoteză nu este satisfăcută.

* **Utilizări ale regresiei polinomiale** (acestea sunt utilizate în principal pentru a defini sau descrie fenomene neliniare), cum ar fi:
  * Rata de creștere a țesuturilor.
  * Progresia epidemiei de boli
  * Distribuția izotopilor de carbon în sedimentele lacurilor

* **y = $w_{0}$ + $w_{1}$ * $x_{1}$ + $w_{2}$ * $x_{1}^2$ + ... + $w_{n}$ * $x_{1}^n$**

* **OBS**: Este liniar datorita coeficientilor **$w_{1}$, $w_{2}$, ... , $w_{n}$**.

  - **y** = variabila dependenta
  - **$x_{1}$, $x_{1}^2$, ... , $x_{1}^n $** = variabilele independente
  - **$w_{1}$, $w_{2}$, ... , $w_{n}$** = coeficienti
  - **$w_{0}$** = bias
