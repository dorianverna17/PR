# **III.** *(OPIONAL)* **Preprocesarea datelor**

## **TEORIE**

### **Ce este preprocesarea datelor?**

* Preprocesarea datelor include paii pe care trebuie s-i urmm pentru a transform sau codific datele, astfel nct s poat fi analizate cu uurin de ctre man.

* Ideea principal pentru c un model s fie precis n predicii este c algoritmul s poat interpreta cu uurin caracteristicile datelor.

* Majoritatea seturilor de date din lumea real pentru ML sunt susceptibile s aib date lipsa, inconsecvene i zgomot din cauza originii lor eterogene.

* Aplicarea algoritmilor de extragere a datelor pe aceste date zgomotoase nu ar da rezultate de calitate, deoarece nu ar reui s identifice modelele n mod eficient. Preprocesarea datelor este, prin urmare, important pentru a mbunti calitatea general a datelor.

* Valorile duplicate sau lips pot oferi o imagine incorect a statisticilor generale ale datelor.

* Valorile aberante i punctele de date inconsecvene tind adesea s perturbe nvarea general a modelului, ceea ce duce la predicii false.

### **Tratarea datelor lips**

* Exist multe motive pentru care datele ar putea lipsi dintr-un set de date. De exemplu, datele colectate printr-un sondaj pot avea date lips din cauza eecului participanilor de a rspunde la unele ntrebri, a nu cunoate rspunsul corect sau c nu doresc s rspund. Pot lipsi i din cauza erorii fcute n timpul procesului de introducere a datelor.

* Majoritatea modelelor de nvare automat necesit date cu o valoare pentru toate caracteristicile din fiecare observaie. n astfel de modele, lipsa datelor poate duce la prtinire n estimarea parametrilor i, de asemenea, poate compromite acurateea modelelor de ML.

#### **Exist mai multe tehnici pe care le folosim pentru a gestiona datele lips, printre care menionm:**

##### 1. **tergerea liniei cu valoarea/valorile lips** (nerecomandat)



  * Aceast tehnic funcioneaz bine pe seturi mari de date cu puine valori lips. De exemplu, tergerea unui rnd dintr-un set de date cu sute de observaii nu poate afecta calitatea informaiilor setului de date. Cu toate acestea, aceast tehnic nu este potrivit pentru un set de date care raporteaz multe valori lips. tergerea multor rnduri dintr-un set de date duce la pierderea informaiilor.

##### 2. **Media** (nerecomandat)



* n cadrul acestei tehnici, nlocuim valoarea lips cu media variabilei n care apare. Avantajul acestei tehnici este c pstreaz media i dimensiunea eantionului. Cu toate acestea, aceast tehnic are unele dezavantaje serioase.

* Imputarea medie subestimeaz eroarea standard i nu pstreaz corelaia dintre variabile. Relaia dintre variabile este un aspect esenial al analizei, deoarece obiectivul general al studiului este de a o nelege mai bine.

* Prin urmare, imputarea mediei nu este o soluie adecvat pentru datele lips dect dac datele lipsesc complet la ntmplare (datele lips nu au nicio legtur att cu datele lips, ct i cu valorile observate din setul de date).

##### 3. **Hot Deck** (recomandat)



* n aceast tehnic, nlocuim valoarea lips a observaiei cu o valoare selectat aleatoriu din toate observaiile din eantion care are valori similare pentru alte variabile.

* Astfel, aceast tehnic asigur c valoarea de imputare este selectat numai din intervalul posibil n care valoarea real ar putea probabil s scad i este selectat aleatoriu, mai degrab dect s fie determinat, ceea ce este un aspect esenial pentru o eroare standard corect.

##### 4. **Cold Deck** (recomandat)



* nlocuim datele lips folosind o valoare aleas dintre alte variabile cu valori similare de observaie n aceast tehnic. Diferena dintre aceast tehnic i imputarea Ho Deck este c procesul de selectare a valorii de imputare nu este randomizat.

##### 5. **Regresie** (recomandat)



* Regresia implic ajustarea unui model de regresie pe o caracteristic cu date lips i apoi utilizarea prediciilor acestui model de regresie pentru a nlocui valorile lips din aceast caracteristic. Aceast tehnic pstreaz relaiile dintre caracteristici i aceast i confer un avantaj semnificativ fa de tehnicile simple de inlocuiew, cum ar fi  media i modulul.

### **Tehnici de preprocesare a datelor pentru ML**

#### **1. Redimensionarea datelor**



* Atunci cnd datele sunt compuse din atribute cu plaje de valori diferite, muli algoritmi de ML pot beneficia de redimensionarea atributelor pentru c toate s aib aceeai scar.

* Acest lucru este util pentru algoritmii de optimizare utilizai n nucleul algoritmilor de ML, cum ar fi gradient descent.

* De asemenea, este util pentru algoritmii care pondereaz intrrile, cum ar fi regresia i reelele neuronale i algoritmii care folosesc msuri de distan, cum ar fi K-Nearest Neighbors.

* Putem redimensiona datele folosind clasa MinMaxScaler.

````python
import pandas
import scipy
import numpy
from sklearn.preprocessing import MinMaxScaler
 
"""
https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database?resource=download

This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. 
The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on 
certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of 
these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.
"""

names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

dataframe = pandas.read_csv("diabetes.csv", names=names)
array = dataframe.values

print("dataframe = ")
print(dataframe.iloc[:10,:])

print("---------------------------------------------------")
print("array = ")
print(array)

# variable independente
X = array[1:,0:8]
# variable dependente
Y = array[1:,8]

print("---------------------------------------------------")
print("X = ")
print(X[:10])

print("---------------------------------------------------")
print("Y = ")
print(Y[:10])
 

scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
 
numpy.set_printoptions(precision=3)

print("---------------------------------------------------")
print("rescaledX = ")
print(rescaledX[0:10,:])
````

#### **3. Standardizarea datelor**

### **Feature Scaling**

**OBS**: Se aplic dup mprirea datelor n antrenare i testare pentru a nu se produce pierdere de informaie nou.

**OBS**: Se folosete pentru c anumite feature-uri s nu fie dominate de altele, le aducem n aceeai plaj de valori. Putem s realizm acest lurcru prin:

  1. **Standardizare** 

  - **$X_{stand}$** = $\frac{X - mean(X)}{standardDeviation(X)}$  [-3, +3]
  - **$standardDeviation(X)$** = $\sqrt{\frac{\sum_{i=1}^{n} (x_{i} - x')^2}{n-1}}$
    - x' = media valorilor lui X
    - n  = nr total de valori

  2. **Normalizare** 

  - **$x_{norm}$** = $\frac{x - min(X)}{max{X}-min(X)}$  [-1, +1]

**OBS** : **Nu** faci feature scaling pe datele de testare.

## **1. Regresie**

### **1. Linear Regression**

#### **TEORIE**

* Regresia modeleaz o valoare de predicie int bazat pe variabile independente. Este folosit mai ales pentru a afl relaia dintre variabile i prognoz.

* Regresia liniar ndeplinete sarcina de a prezice o valoare a variabilei dependene (y) pe baza unei variabile independente date (x). Deci, aceast tehnic de regresie descoper o relaie liniar ntre x (intrare) i y (ieire).

* **y = $w_{0}$ + $w_{1}$ * $x$**

  - **y** = variabil dependen
  - **x** = variabil independena
  - **$w_{1}$** = coeficient
  - **$w_{0}$** = bias

* Cnd antrenm modelul  potrivim cea mai bun linie pentru a prezice valoarea lui y pentru o valoare dat a lui x. Modelul obine cea mai bun linie de potrivire de regresie prin gsirea celor mai bune valori **$w_{0}$** i **$w_{1}$**.

* Cum se actualizeaz valorile **$w_{0}$** i **$w_{1}$** pentru a obine cea mai bun linie de regresie?

* Prin realizarea liniei de regresie cea mai bun, modelul urmrete s prezic valoarea y astfel nct diferena de eroare dintre valoarea prezis i valoarea adevrat s fie minim. Deci, este foarte important s actualizm valorile  **$w_{0}$** i **$w_{1}$**, pentru a ajunge la cea mai bun valoare care s minimizeze eroarea dintre valoarea y prezis (pred) i valoarea y adevrat (y).

  - **$RMSE$** = $\frac{1}{n}$ * ${\sum_{i=1}^{n} (pred_{i} - y_{i})^2}$

  - **minimize** **$RMSE$**

* Root Mean Squared Error (**$RMSE$**)

* Pentru a actualiza valorile **$w_{0}$** i **$w_{1}$** pentru a reduce funcia Cost (minimizarea valorii RMSE) i pentru a obine cea mai bun linie de regresie, modelul folosete **Gradient Descent** (https://www.geeksforgeeks.org/gradient-descent-n-linear-regression/?ref=rp) . Ideea este de a ncepe cu valori aleatoare ale **$w_{0}$** i **$w_{1}$** i apoi actualizarea iterativ a valorilor, ajungnd la costul minim.

* **Exemplu**: Relaia dintre nivelurile de poluare i creterea temperaturii.

### **3. Random Forest Regression**

#### **TEORIE**

* Fiecare arbore de decizie are o varian mare, dar cnd le combinm pe toate mpreun n paralel, variana rezultat este sczut, deoarece fiecare arbore de decizie este antrenat perfect pe acele date eantion, i, prin urmare, rezultatul nu depinde de un arbore de decizie, ci de arbori de decizie multipli. n cazul unei probleme de clasificare, rezultatul final este luat prin utilizarea clasificatorului cu vot majoritar. n cazul unei probleme de regresie, rezultatul final este media tuturor rezultatelor. Aceast parte se numete agregare.

* Random Forest este o ensemble technique capabil s efectueze att sarcini de regresie, ct i de clasificare cu ajutorul mai multor arbori de decizie i o tehnic numit **Bootstrap & Aggregation**, cunoscut n mod obinuit c **bagging**. Ideea de baz din spatele acestui lucru este de a combin mai muli arbori de decizie n determinarea rezultatului final, mai degrab dect s se bazeze pe arbori de decizie individuali.
Random Forest are mai muli arbori de decizie c modele de baz de nvare. Efectum aleatoriu eantionarea rndurilor i eantionarea caracteristicilor din setul de date, formnd seturi de date mostre pentru fiecare model. Aceast parte se numete **Bootstrap**.

* **EXEMPLU**: Random Forest este folosit att pentru clasificare, ct i pentru regresie - de exemplu, pentru clasificarea dac un e-mail este spam sau nu spam. Random Forest este utilizat n multe industrii diferite, inclusiv bancar, comer i asisten medical.

### **4. Multiple Linear Regression**

#### **TEORIE**

* Regresia este o metod statistic pentru a determina relaia dintre caracteristici i o variabil sau rezultat. nvarea automat, este utilizat c metod de modelare predictiv, n care se folosete un algoritm pentru a prognoza rezultate continue. Regresia liniar multipl, cunoscut adesea sub numele de regresie multipl, este o metod statistic care prezice rezultatul unei variabile de rspuns prin combinarea a numeroase variabile explicative. Regresia multipl este o variant a regresiei liniare n care este utilizat o singur variabil explicativ.

*  **y = $w_{0}$ + $w_{1}$ * $x_{1}$ + $w_{2}$ * $x_{2}$ + ... + $w_{n}$ * $x_{n}$**

  - **y** = variabil dependen
  - **$x_{1}$, $x_{2}$, ... , $x_{n}$** = variabilele independente
  - **$w_{1}$, $w_{2}$, ... , $w_{n}$** = coeficieni
  - **$w_{0}$** = bias

* **EXEMPLU**: Un analist poate dori s tie cum micarea pieei afecteaz preul ExxonMobil (XOM). n acest caz, ecuaia lor liniar va avea valoarea indicelui S&P 500 c variabil independent, sau predictor, i preul XOM c variabil dependent.

### **5. Polynomial Regression**

#### **TEORIE**


* Regresia polinomial este o form de regresie liniar n care relaia dintre variabila independent x i variabila dependent y este modelat ca un polinom de gradul al n-lea. Regresia polinomial se potrivete unei relaii neliniare ntre valoarea lui x i media condiionat corespunztoare a lui y, notat $E(y|x)$

* De ce regresia polinomial?
 
  * Exist unele relaii despre care un cercettor le va ipoteza c sunt curbilinii. n mod clar, astfel de tipuri de cazuri vor include un termen polinom.Verificarea reziduurilor. Dac ncercm s potrivim un model liniar la datele curbe, un grafic de mprtiere a reziduurilor (axa Y) pe predictor (axa X) va avea pete cu multe reziduuri pozitive n mijloc. Prin urmare, ntr-o astfel de situaie, nu este potrivit.
  * O ipotez n analiza de regresie liniar multipl obinuit este c toate variabilele independente sunt independente. n modelul de regresie polinomial, aceast ipotez nu este satisfcut.

* **Utilizri ale regresiei polinomiale** (acestea sunt utilizate n principal pentru a defini sau descrie fenomene neliniare), cum ar fi:
  * Rata de cretere a esuturilor.
  * Progresia epidemiei de boli
  * Distribuia izotopilor de carbon n sedimentele lacurilor

* **y = $w_{0}$ + $w_{1}$ * $x_{1}$ + $w_{2}$ * $x_{1}^2$ + ... + $w_{n}$ * $x_{1}^n$**

* **OBS**: Este liniar datorita coeficientilor **$w_{1}$, $w_{2}$, ... , $w_{n}$**.

  - **y** = variabila dependenta
  - **$x_{1}$, $x_{1}^2$, ... , $x_{1}^n $** = variabilele independente
  - **$w_{1}$, $w_{2}$, ... , $w_{n}$** = coeficienti
  - **$w_{0}$** = bias
