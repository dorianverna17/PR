## **1.1 Similariti ntre reelele neurale artificiale i cele biologice, din creierul uman**

## **1.2 Structura general a unei reele neurale artificiale**


* O reea neurala este organizat pe straturi de neuroni. Aceste straturi de neuroni sunt de trei feluri: strat de intrare, strat de ieire i strat ascuns. Straturile de intrare i ieire reprezint modul prin care putem folosi modelul. Dac avem un model ce ndeplinete o anumit sarcina prin a procesa un set de date, aceste date sunt trecute prin model ncepnd cu stratul de intrare i rezultatul este reflectat de valorile produse de stratul de ieire. n cazul straturilor ascunse, fiecare neuron de pe fiecare strat este conectat la toi neuronii de pe stratul precedent i de pe stratul urmtor.

* **Termeni de reinut**

  1.   Toate "**legturile**" dintre neuronii unei reele neurale artificiale reprezint un **"weight"/"pondere"**. Aceast pondere este nmulit cu datele ce "**intr**" n acel neuron i este apoi utilizat mai departe.

  2.   Informaia ce "**iese**" dintr-un neuron este rezultatul trecerii datelor de intrare printr-o ***funcie de activare***.


## **1.3 Cum funcioneaz o reea neurala din punct de vedere matematic?**

* Operaiile matematice ce stau la baza reelelor neurale artificiale sunt relativ simple. Putem lua c exemplu urmtoarea reea neural:

neural-network-1.png

* Datele de intrare sunt reprezentate de vectorul:

\begin{align} X = \begin{bmatrix}
           x_{1} \\
           x_{2} \\
\end{bmatrix} \end{align}

* Ponderile sunt reprezentate de matricea:

\begin{align} W = \begin{bmatrix}
           w_{1} & w_{2} \\
\end{bmatrix} \end{align}

* Rezultatul Z va avea urmtoarea formul: 

$$ Z = F(WX+b) $$
\begin{align}
        W \cdot X = \sum_{i=1}^n w_i x_i.
\end{align}

* Unde F este funcia de activare i b este un bias.

## **1.4 Ce este o funcie de activare?**

* n termeni simpli, este o funcie ce ajut o reea neural s nvee s modeleze date complexe. Dac am face o analogie ntre reelele neurale artificiale i reelele neurale biologice, din interiorul creierelor, funcia de activare reprezint un element ce decide dac informaia s fie transmis mai departe la urmtorul neuron.

* **Exemplu de funcie de activare**

* n poz de mai sus avem o reea neural puin mai complicat c cea precedent, cu un strat de intrare, un strat ascuns i un strat de ieire.

* C date de intrare avem vectorul 
\begin{align} X = \begin{bmatrix}
           x_{1} \\
           x_{2} \\
\end{bmatrix} \end{align}

* Ca date de ieire avem vectorul $A_{2}$ din neuronul din stratul de ieire.

* Avnd un strat ascuns, vom avea dou matrici de ponderi (W1 i W2). Una intr stratul de intrare i cel ascuns i una ntre stratul ascuns i cel de ieire.
Acestea pot fi scrise n modul urmtor:
\begin{align} W1 = \begin{bmatrix}
           w_{11} & w_{21} \\
           w_{12} & w_{22} \\
           w_{13} & w_{23} \\
         \end{bmatrix} 
         \end{align}

\begin{align} W2 = \begin{bmatrix}
           w_{11} & w_{21}\\
         \end{bmatrix} 
\end{align}
        
* Atunci cnd dorim s folosim modelul, vectorul X este nmulit cu W1 i se va obine:
\begin{align} Z = \begin{bmatrix}
           w_{11} * x_{1} + w_{21} * x_{2} \\
           w_{12} * x_{1} + w_{22} * x_{2}\\
           w_{13} * x_{1} + w_{23} * x_{2}\\
         \end{bmatrix} \end{align}

* Adunm apoi bias-ul $b1$ i vom obine vectorul:
\begin{align} Z + b_{1} = \begin{bmatrix}
           w_{11} * x_{1} + w_{21} * x_{2} + b_{1}\\
           w_{12} * x_{1} + w_{22} * x_{2} + b_{1}\\
           w_{13} * x_{1} + w_{23} * x_{2} + b_{1}\\
         \end{bmatrix} \end{align}
        
* Fiecare element din acest vector reprezint datele de intrare pentru fiecare neuron de pe stratul ascuns. Aplicnd apoi o funcie de activare $F_{1}$ pe datele de intrare, vom obine vectorul $A_{1}$, ce reprezint datele de ieire din stratul ascuns.

## **Crearea reelei neurale**

* Crearea unei reele neurale artificiale n Keras este destul de simpl:

```python
batch_size = 64
epochs = 10

# Crearea unui model in Keras
model = Sequential()

# Adaugarea unui strat de intrare ce primeste un vector de 784 elemente
model.add(InputLayer(input_shape=(784,)))

# Adaugarea mai multor straturi ascunse
# Observati functia de activare a fiecarui strat, "relu" (formula matematica o gasiti in poza de mai sus cu functii de activare)
model.add(Dense(64, activation="relu"))
model.add(Dense(128, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))

# Adaugarea stratului de iesire
# Observati ca functia de activare este "softmax", ce asigura ca output-ul va fi un vector
# de probabilitati, cu suma elementelor egala cu 1.
model.add(Dense(10, activation="softmax"))
```

```python
# Pentru a "finaliza" crearea unui model, acesta trebuite "compilat"
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Inainte de a antrena modelul trebuie sa ne asiguram ca seturile de date sunt NumPy arrays
X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)
```

* Antrenarea unei reele neurale artificiale, pentru a obine un model ct mai bun, ntr-un timp ct mai scurt, se face n mai multe **batch-uri** i n mai multe **epoci**.

* Un batch de mrime N reprezint o bucat de N elemente din setul de date ce este folosit pentru antrenarea modelului i dup care sunt actualizate ponderile reelei neurale. Antrenarea unei reele neurale n batch-uri are urmtoarele avantaje:

  1.   Necesit mai pun memorie RAM. Dac avem un set de date foarte mare, e foarte posibil s nu l putem ncarc pe tot n memorie.
  2.   Reeaua se antreneaz mai repede, pentru c actualizarea iterativ a ponderilor se face mai des.

* O epoca reprezint o trecere complet prin setul de date. Acest proces este similar cu nvatul pentru un examen, cnd citeti toat teoria de la cap la coad de mai multe ori pentru a o reine ct se poate de bine. O reea neurala procedeaz la fel pentru a nva mai bine, trecnd de mai multe ori prin ntregul set de date.

```python
batch_size = 64
epochs = 10

# Antrenam reteaua creata mai devreme
# Observam parametrul "verbose", folosit pentru a obtine progresul modelului.
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose = 1)
```

* Prin rularea celulei precedente putem observa cum acurateea crete de la o epoc la alta, dar se mbuntete din ce n ce mai pun cu fiecare epoc.

* ntre prima epoc i a doua este o mbuntire de 10% a acuratetei, n timp ce ntre a 7-a i a 8-a este o mbuntire de 0.25% i ntre a 15-a i a 16-a de 0.02%. Pentru fiecare problema n parte i pentru fiecare model exist un numr de epoci maxim pentru care merit s fie antrenat modelul. Odat ce acel numr de epoci este depit, mbuntirea este neglijabil.

* Cea mai simpl metod de a evalua performan unui model pe setul de testare este aceea de a determina acurateea sa. n keras, avem funcia "**evaluate**" a modelului:

```python
_, accuracy = model.evaluate(X_test, y_test)
print('Accuracy: %.2f' % (accuracy*100))
```

* Modelul creat de noi a obinut, pe un set de date nemaivzut, o acuratee de peste 95%, deci am putea spune c are o performan foarte bun. Dac am ncerca s prezicem n mod complet aleatoriu ce cifra se afl n fiecare imagine din setul de testare, din punct de vedere statistic, ar trebui s prezicem corect n 10% din cazuri. Modelul nostru aduce o mbuntire de 85%, deci am putea spune c are o performan foarte bun.

## **5.2 Cross-validation**

* Setul de date MNIST este un caz fericit, deoarece avem datele mprite deja ntr-un set de testare i un set de antrenare i avem i un numr destul de mare de elemente n set. n practic un caz foarte des ntlnit este acela de a nu avea multe date cu care s putem lucra, ceea ce poate duce foarte des la modele cu o performan sczut.

* O soluie la aceast problema se numete **cross-validation**. Aceasta este o metod de antrenare a unui model de inteligen artificial atunci cnd avem un set de antrenare de dimensiune redus, dar care totui poate modela destul de bine problema ce trebuie rezolvat. n romn, aceast metod s-ar traduce "validare n cruce" (cred).

* Validarea n cruce presupune mprirea setului de date n N buci i antrenarea modelului pe N-1 dintre acestea, urmnd c pe bucat rmas s se calculeze performan modelului pentru a putea fi actualizate ponderile reelei. Acest proces are loc de N ori, pn cnd fiecare bucat este folosit pentru a calcula performan modelului.
