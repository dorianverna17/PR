# **I. Cum funcționează o rețea neurala?**



## **1.1 Similarități între rețelele neurale artificiale și cele biologice, din creierul uman**

## **1.2 Structura generală a unei rețele neurale artificiale**


* O rețea neurala este organizată pe straturi de neuroni. Aceste straturi de neuroni sunt de trei feluri: strat de intrare, strat de ieșire și strat ascuns. Straturile de intrare și ieșire reprezintă modul prin care putem folosi modelul. Dacă avem un model ce îndeplinește o anumită sarcina prin a procesa un set de date, aceste date sunt trecute prin model începând cu stratul de intrare și rezultatul este reflectat de valorile produse de stratul de ieșire. În cazul straturilor ascunse, fiecare neuron de pe fiecare strat este conectat la toți neuronii de pe stratul precedent și de pe stratul următor.

* **Termeni de reținut**

  1.   Toate "**legăturile**" dintre neuronii unei rețele neurale artificiale reprezintă un **"weight"/"pondere"**. Această pondere este înmulțită cu datele ce "**intră**" în acel neuron și este apoi utilizată mai departe.

  2.   Informația ce "**iese**" dintr-un neuron este rezultatul trecerii datelor de intrare printr-o ***funcție de activare***.


## **1.3 Cum funcționează o rețea neurala din punct de vedere matematic?**

* Operațiile matematice ce stau la baza rețelelor neurale artificiale sunt relativ simple. Putem lua că exemplu următoarea rețea neurală:

neural-network-1.png

* Datele de intrare sunt reprezentate de vectorul:

\begin{align} X = \begin{bmatrix}
           x_{1} \\
           x_{2} \\
\end{bmatrix} \end{align}

* Ponderile sunt reprezentate de matricea:

\begin{align} W = \begin{bmatrix}
           w_{1} & w_{2} \\
\end{bmatrix} \end{align}

* Rezultatul Z va avea următoarea formulă: 

$$ Z = F(WX+b) $$
\begin{align}
        W \cdot X = \sum_{i=1}^n w_i x_i.
\end{align}

* Unde F este funcția de activare și b este un bias.

## **1.4 Ce este o funcție de activare?**

* În termeni simpli, este o funcție ce ajută o rețea neurală să învețe să modeleze date complexe. Dacă am face o analogie între rețelele neurale artificiale și rețelele neurale biologice, din interiorul creierelor, funcția de activare reprezintă un element ce decide dacă informația să fie transmisă mai departe la următorul neuron.

* **Exemplu de funcție de activare**



Commonly-used-activation-functions-a-Sigmoid-b-Tanh-c-ReLU-and-d-LReLU.png

crop_NN.jpg

* În poză de mai sus avem o rețea neurală puțin mai complicată că cea precedentă, cu un strat de intrare, un strat ascuns și un strat de ieșire.

* Că date de intrare avem vectorul 
\begin{align} X = \begin{bmatrix}
           x_{1} \\
           x_{2} \\
\end{bmatrix} \end{align}

* Ca date de ieșire avem vectorul $A_{2}$ din neuronul din stratul de ieșire.

* Având un strat ascuns, vom avea două matrici de ponderi (W1 și W2). Una intră stratul de intrare și cel ascuns și una între stratul ascuns și cel de ieșire.
Acestea pot fi scrise în modul următor:
\begin{align} W1 = \begin{bmatrix}
           w_{11} & w_{21} \\
           w_{12} & w_{22} \\
           w_{13} & w_{23} \\
         \end{bmatrix} 
         \end{align}

\begin{align} W2 = \begin{bmatrix}
           w_{11} & w_{21}\\
         \end{bmatrix} 
\end{align}
        
* Atunci când dorim să folosim modelul, vectorul X este înmulțit cu W1 și se va obține:
\begin{align} Z = \begin{bmatrix}
           w_{11} * x_{1} + w_{21} * x_{2} \\
           w_{12} * x_{1} + w_{22} * x_{2}\\
           w_{13} * x_{1} + w_{23} * x_{2}\\
         \end{bmatrix} \end{align}

* Adunăm apoi bias-ul $b1$ și vom obține vectorul:
\begin{align} Z + b_{1} = \begin{bmatrix}
           w_{11} * x_{1} + w_{21} * x_{2} + b_{1}\\
           w_{12} * x_{1} + w_{22} * x_{2} + b_{1}\\
           w_{13} * x_{1} + w_{23} * x_{2} + b_{1}\\
         \end{bmatrix} \end{align}
        
* Fiecare element din acest vector reprezintă datele de intrare pentru fiecare neuron de pe stratul ascuns. Aplicând apoi o funcție de activare $F_{1}$ pe datele de intrare, vom obține vectorul $A_{1}$, ce reprezintă datele de ieșire din stratul ascuns.

# **II. Cum funcționează procesul de învățare al unei rețele neurale?**

## **2.1 Funcția de cost**

* Funcția de cost reprezintă o funcție ce determină cât de mare este eroarea dintre rezultatul generat de model și rezultatul adevărat ce ar fi trebuit să fie generat de model. Această funcție duce 
În procesul de antrenare a unei rețele neurale artificiale, rețeaua începe cu toate ponderile inițializate aleatoriu. Aceste ponderi sunt apoi modificate pe parcursul procesului de antrenare prin a se calcula eroarea cu funcția de cost și, folosind această eroare și procesul de **backpropagation**, ponderile sunt corectate în mod iterativ.

## **2.2 Backpropagation**


* Procesul de backpropagation identifică ce "rute" în rețea "influențează" cel mai mult rezultatul pentru a le putea crește sau scădea influența că să ajungem la rezultatul dorit.

# **III. Antrenarea unei rețele neurale în practică**

## **3.1 Obiectivul principal**


* Atunci când dorim să antrenăm un model de inteligență artificială, obiectivul nostru nu este doar acela de a crea un model ce reușește să facă o corelare foarte bună între datele de intrare și cele de ieșire din setul de date pe care îl folosim la antrenare. Scopul nostru principal este acela de a crea un model care reușește să înțeleagă problema îndeajuns de bine încât să poată generaliza corelarea dintre datele de intrare și ieșire și pe niște date pe care nu le-a mai întâlnit.

* Spre exemplu, atunci când înveți operația de adunare la școală, începi prin a învață că $1+1=2$ și $2+2=4$, dar prin exerciții repetate, înveți să generalizezi și să rezolvi și o adunare pe care nu ai mai întâlnit-o, cum ar fi $2968 + 1078$.

## **3.2 Punere în practică**


* Crearea unui model ce este capabil să generalizeze și pe date pe care nu le-a mai văzut pare o sarcină destul de dificilă, deoarece cel mai bun mod de a îi evalua performanța este acela de a-l folosi pe date din viață reală, nu doar din setul nostru de date. Pentru a simula această situație putem împărți setul nostru de date într-un set de antrenare și unul de testare, în așa fel încât niciun element din setul de testare nu se găsește în setul de antrenare. Astfel, după ce modelul este antrenat, putem să îl evaluăm pe date pe care nu le-a mai văzut. De obicei, setul de antrenare conține 80-90% din setul de date inițial și cel de testare restul de 10-20%.

## **Crearea rețelei neurale**

* Crearea unei rețele neurale artificiale în Keras este destul de simplă:

```python
batch_size = 64
epochs = 10

# Crearea unui model in Keras
model = Sequential()

# Adaugarea unui strat de intrare ce primeste un vector de 784 elemente
model.add(InputLayer(input_shape=(784,)))

# Adaugarea mai multor straturi ascunse
# Observati functia de activare a fiecarui strat, "relu" (formula matematica o gasiti in poza de mai sus cu functii de activare)
model.add(Dense(64, activation="relu"))
model.add(Dense(128, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))

# Adaugarea stratului de iesire
# Observati ca functia de activare este "softmax", ce asigura ca output-ul va fi un vector
# de probabilitati, cu suma elementelor egala cu 1.
model.add(Dense(10, activation="softmax"))
```

```python
# Pentru a "finaliza" crearea unui model, acesta trebuite "compilat"
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Inainte de a antrena modelul trebuie sa ne asiguram ca seturile de date sunt NumPy arrays
X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)
```

* Antrenarea unei rețele neurale artificiale, pentru a obține un model cât mai bun, într-un timp cât mai scurt, se face în mai multe **batch-uri** și în mai multe **epoci**.

* Un batch de mărime N reprezintă o bucată de N elemente din setul de date ce este folosită pentru antrenarea modelului și după care sunt actualizate ponderile rețelei neurale. Antrenarea unei rețele neurale în batch-uri are următoarele avantaje:

  1.   Necesită mai puțînă memorie RAM. Dacă avem un set de date foarte mare, e foarte posibil să nu îl putem încarcă pe tot în memorie.
  2.   Rețeaua se antrenează mai repede, pentru că actualizarea iterativă a ponderilor se face mai des.

* O epoca reprezintă o trecere completă prin setul de date. Acest proces este similar cu învățatul pentru un examen, când citești toată teoria de la cap la coadă de mai multe ori pentru a o reține cât se poate de bine. O rețea neurala procedează la fel pentru a învață mai bine, trecând de mai multe ori prin întregul set de date.

```python
batch_size = 64
epochs = 10

# Antrenam reteaua creata mai devreme
# Observam parametrul "verbose", folosit pentru a obtine progresul modelului.
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose = 1)
```

* Prin rularea celulei precedente putem observa cum acuratețea crește de la o epocă la alta, dar se îmbunătățește din ce în ce mai puțîn cu fiecare epocă.

* Între prima epocă și a doua este o îmbunătățire de 10% a acuratetei, în timp ce între a 7-a și a 8-a este o îmbunătățire de 0.25% și între a 15-a și a 16-a de 0.02%. Pentru fiecare problema în parte și pentru fiecare model există un număr de epoci maxim pentru care merită să fie antrenat modelul. Odată ce acel număr de epoci este depășit, îmbunătățirea este neglijabilă.

* Cea mai simplă metodă de a evalua performanță unui model pe setul de testare este aceea de a determina acuratețea sa. În keras, avem funcția "**evaluate**" a modelului:

```python
_, accuracy = model.evaluate(X_test, y_test)
print('Accuracy: %.2f' % (accuracy*100))
```

* Modelul creat de noi a obținut, pe un set de date nemaivăzut, o acuratețe de peste 95%, deci am putea spune că are o performanță foarte bună. Dacă am încerca să prezicem în mod complet aleatoriu ce cifra se află în fiecare imagine din setul de testare, din punct de vedere statistic, ar trebui să prezicem corect în 10% din cazuri. Modelul nostru aduce o îmbunătățire de 85%, deci am putea spune că are o performanță foarte bună.

## **5.1 Model fine-tuning**

* Atunci când antrenăm o rețea neurală, această poate fi definită pe scurt de arhitectura sa. În keras, această poate fi vizualizată în modul următor:

* În arhitectura modelului nostru, observăm un număr de 69226 parametri ce pot fi antrenați. Aceștia cuprind ponderile din întreaga rețea, bias-urile, etc. Parametrii ce nu pot fi antrenați se numesc hyper-parametri. Aceștia pot fi numărul de epoci, mărimea unui batch, numărul de neuroni din fiecare strat, funcțiile de activare, etc.

* Când creăm un model de inteligență artificială, de multe ori suntem nevoiți să încercăm mai multe configurații de hyper-parametri pentru a ajunge la o configurație cât se poate de bună. Astfel, în acest proces, putem obține mai multe modele, fiecare cu propria să arhitectură. Pentru a verifică performanță fiecărui model obținut, putem folosi un **set de validare**, obținut prin a împărți setul de antrenare în două seturi, unul strict pentru antrenare și unul pentru validare. Astfel, obținem 3 seturi de date:

  1.   Set de antrenare -> pentru a antrena modelele.
  2.   Set de validare -> pentru a evalua fiecare model și a le compara între ele.
  3.   Set de testare -> Pentru a verifica performanța celui mai bun model obținut pe un set de date complet diferit, nemaivăzut până acum de niciun model.



## **5.2 Cross-validation**

* Setul de date MNIST este un caz fericit, deoarece avem datele împărțite deja într-un set de testare și un set de antrenare și avem și un număr destul de mare de elemente în set. În practică un caz foarte des întâlnit este acela de a nu avea multe date cu care să putem lucra, ceea ce poate duce foarte des la modele cu o performanță scăzută.

* O soluție la această problema se numește **cross-validation**. Aceasta este o metodă de antrenare a unui model de inteligență artificială atunci când avem un set de antrenare de dimensiune redusă, dar care totuși poate modela destul de bine problema ce trebuie rezolvată. În română, această metodă s-ar traduce "validare în cruce" (cred).

* Validarea în cruce presupune împărțirea setului de date în N bucăți și antrenarea modelului pe N-1 dintre acestea, urmând că pe bucată rămasă să se calculeze performanță modelului pentru a putea fi actualizate ponderile rețelei. Acest proces are loc de N ori, până când fiecare bucată este folosită pentru a calcula performanță modelului.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

## **Exercițiul 6 (2 puncte)**

* Creați un model nou, cu o arhitectură pe care o puteți alege cum doriți.

```python
# Creati modelul
model = Sequential()

# Adaugati primul strat, cel de input, cu un input_shape de (11,)
model.add(InputLayer(input_shape=(11,)))

# Adaugati mai multe straturi Dense, cu functia de activare 'relu'.
model.add(Dense(64, activation="relu"))
model.add(Dense(128, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))

# Adaugati stratul de iesire, cu un singur neuron si functia de activare 'sigmoid'.
model.add(Dense(1, activation="sigmoid"))

# Compilam modelul
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

* În cazul altor seturi de date este foarte posibil să nu se repete această situație. Pentru a putea prezenta cum se implementează Cross-Validation, vom presupune că avem nevoie de Cross-Validation, chiar dacă nu avem.

```python
from sklearn.model_selection import KFold

# KFold reprezinta structura de date care va imparti setul nostru in N portiuni
N = 4
# train_index si test_index reprezinta elementele selectate pentru antrenare si testare la un singur fold
kf = KFold(n_splits=4)

for train_index, test_index in kf.split(X):
  # Cream seturile de antrenare si testare din setul de date original
  x_train, x_test = X[train_index], X[test_index]
  y_train, y_test = y[train_index], y[test_index]

  # Antrenam modelul
  model.fit(x_train, y_train, epochs=1, batch_size=batch_size, verbose = 1)


  # Testam modelul
  _, accuracy = model.evaluate(x_test, y_test)
  print('Accuracy: %.2f' % (accuracy*100))
```
